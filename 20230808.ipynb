1. 가상환경 구축
    conda env list -> 기존 가상환경 목록 확인하기

    conda create -n prjname python=3.9 ipython numpy pandas matplotlib scipy

    -> 가상환경 설치

    -> (y/n) y

    -> 설치완료 후 결과
        conda activate prjname 문구 확인


    streamlit은 오류 발생하는 경우가 있어 가상환경 구축 이후에 하는게 좋음

    -> conda activate prjname
    가상환경 진입
    (프롬포트 앞에 (prjname)이 나오는지 확인)
    나가고 싶을땐 가상환경 종료 ->
    conda deactivate

    -> python --version  
    파이썬 버전 확인

    -> python -> import numpy, pandas, matplotlib 잘 되었는지 확인
