```python
import glob

def FileCollection(L_name, file_types):

    s_path = '/'.join(str(__import__(L_name)).split()[-1].strip('\'').split('/')[:-1]) + '/'
    #s_path = '\\'.join(str(__import__(L_name)).split()[-1].strip('\'').split('\\')[:-1]) + '\\'    Window 로컬에서 경로명을 얻어내는 방법
    #s_path = str(__import__(L_name)).split('\'')[3][:-11]
    #s_path = '\\'.join(s.split()[-1].split('\\')[:-1])+'\\'
    print(s_path)

    dummy_arg = s_path + '**/*.'+file_types
    #dummy_arg = s_path + '**\\\*.' + file_types
    print(dummy_arg)
    file_lists = glob.glob(dummy_arg, recursive=True)
    return (file_lists, s_path)

def GetFileSentence(file_list, search_sentences, s_path):
    file_dict = {}
    for filename in file_list:
        #print(filename)
        sentence_list = []
        with open(filename, encoding = 'utf-8') as f:
            lines = f.readlines()
        for item in lines:
            if search_sentences in item:
                sentence_list.append(item)
        if sentence_list :
            file_dict[filename[len(s_path):]] = sentence_list
    return(file_dict)

def print_result(fdr):
    for i in fdr:
        print(i)
        print('==='*10)
        for j in fdr[i]:
            print('-   ', j.strip())
        print()
```

```python
(file_list , s_path) = FileCollection('numpy', 'py')
f_dict = GetFileSentence(file_list, 'index', s_path)
print_result(f_dict)
```

```python
def brute_force_algorithm(total_head, total_leg):
    result = []
    for pig in range(total_head + 1):
        for chicken in range(total_head - pig + 1):
            bug = total_head - pig - chicken
            if 2*pig + 4*chicken + 8*bug == total_leg:
                result.append((pig, chicken, bug))
    return(result)
```

```python
def barnyard():
    total_head = int(input('머리의 총 합을 입력하세요.:'))
    total_leg = int(input('다리의 총 합을 입력하세요.:'))

    result = brute_force_algorithm(total_head,total_leg)

    if result :
        for i in result:
            print(f'닭 수: {i[0]}, 돼지 수 : {i[1]}, 거미 수 : {i[2]}')
    else:
        print('There is no solution')
```

```python
def ramanuzan(n):
    result = []
    for i in range(n+1):
        for j in range(n+1):
            if i**3 + j**3 == n:
                if i < j:
                    result.append((i,j))
    return result
```
